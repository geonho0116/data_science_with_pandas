### 결측치

```python
df.isnull()
```

```python
null_count = df.isnull().sum()
```

```python
null_count.plot.barh(figsize=(6,8))
```

```python
# 위에서 계산한 결측치 수를 reset_index를 통해 데이터프레임으로 만든다.
# df_null_count 변수에 결과를 담아서 head로 미리보기를 한다.

df_null_count = null_count.reset_index()
df_null_count.head()
```

### 컬럼명 변경하기

```python
df_null_count.columns = ["컬럼명","결측치수"]
```

```python
df_null_count.head()
```

### 정렬하기

```python
# df_null_count 데이터프레임에 있는 결측치 수 컬럼을 sort_values를 통해 정렬하여
# 결측치가 많은 순으로 상위 10개만 출력한다.

df_null_count_top = df_null_count.sort_values(by="결측치수", ascending=False).head(10)
```

### 특정 컬럼만 불러오기

```python
#NaN = Not a Number 의 약자로 결측치를 의미한다.

df["지점명"].head()
```

```python
drop_columns=df_null_count_top["컬럼명"].tolist()
drop_columns
```

```python
df[drop_columns].head()
```

### 제거하기

```python
# axis=0 행 기준 axis=1 열 기준
print(df.shape)
df = df.drop(drop_columns, axis=1)
print(df.shape)
```

```python
df.info()
```
